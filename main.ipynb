# Popularity Based System

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean(numeric_only = True)['Book-Rating'].reset_index()

avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author','Year-Of-Publication',	'Publisher',	'Image-URL-M', 'num_ratings', 'avg_ratings']].reset_index()
```

# Collaborative Filtering Based Recommender System

```python
z = ratings_with_name.groupby('User-ID').count()
z = z[z['Book-Rating']>200]
```

```python
req_users = z.index
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(req_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt = pt.fillna(0.0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:20]
    for i in similar_items:
        print(pt.index[i[0]])
    return
```

```python
recommend('Sanctuary')
```
